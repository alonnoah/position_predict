```python
all_data = pd.read_csv("AllTheData.csv")
clean_data = all_data.copy()
clean_data_2 = all_data.copy()
all_data.columns   # name's of the colums in the daa=ta
```

```python
for ind,pos in enumerate(clean_data['Preferred Positions']):
    if (pos.startswith('LB') or pos.startswith('RB') or pos.startswith('LWB')):
        clean_data.loc[ind, 'Preferred Positions'] = 1
    elif (pos.startswith('SW') or pos.startswith('RWB') or pos.startswith('CB')):
        clean_data.loc[ind, 'Preferred Positions'] = 1
    elif (pos.startswith('CAM') or pos.startswith('CM') or pos.startswith('CDM')):
        clean_data.loc[ind, 'Preferred Positions'] = 2
    elif (pos.startswith('RM') or pos.startswith('LM')):
        clean_data.loc[ind, 'Preferred Positions'] = 2
    elif (pos.startswith('LW') or pos.startswith('LF') or pos.startswith('ST')):
        clean_data.loc[ind, 'Preferred Positions'] = 3
    elif (pos.startswith('CF') or pos.startswith('RF') or pos.startswith('RW')):
        clean_data.loc[ind, 'Preferred Positions'] = 3
    elif(pos.startswith('GK')):
           clean_data.drop(ind, axis=0, inplace=True)
```

```python
clean_data['Preferred Foot'].replace({"Right": 0, "Left": 1}, inplace=True)
clean_data.to_csv(r'/Users/user/PycharmProjects/DataLeNetunim/eda_2_data.csv', index=False)
```
